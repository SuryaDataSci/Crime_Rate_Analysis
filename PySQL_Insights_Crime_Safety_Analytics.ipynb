```python
# Clean column names to remove spaces and ensure consistency
df.columns = df.columns.str.strip().str.lower()  # Optionally, convert to lowercase

# Now load it into SQL again
df.to_sql('crime_records', con=engine, if_exists='replace', index=False)
```

```python
#now aggregating the rows and columns with the above specified rows
# SQL query to aggregate crimes by Area_Name, Group_Name, and Year
query_aggregate = """
    SELECT Area_Name AS City, Group_Name AS Crime_Type, Year, 
           SUM(Cases_Property_Stolen) AS Total_Crimes_Stolen,
           SUM(Cases_Property_Recovered) AS Total_Crimes_Recovered
    FROM crime_records
    GROUP BY Area_Name, Group_Name, Year
    ORDER BY Total_Crimes_Stolen DESC;
"""

# Execute query and store the result in a DataFrame
df_summary = pd.read_sql(query_aggregate, con=engine)

# Display the result
print(df_summary.head())
```

```python
#now Classify Crimes Based on Severity Using CASE Statements
# SQL query to classify crimes based on severity
query_severity = """
    SELECT Area_Name AS City, Group_Name AS Crime_Type, Year, 
           CASE 
               WHEN Group_Name = 'Murder' THEN 10
               WHEN Group_Name = 'Rape' THEN 8
               WHEN Group_Name = 'Theft' THEN 3
               WHEN Group_Name = 'Robbery' THEN 7
               ELSE 1
           END AS Severity_Score,
           SUM(Cases_Property_Stolen) AS Total_Crimes_Stolen
    FROM crime_records
    GROUP BY Area_Name, Group_Name, Year
    ORDER BY Severity_Score DESC;
"""
# Execute query and store the result in a DataFrame
df_severity = pd.read_sql(query_severity, con=engine)

# Display the result
print(df_severity.head())
```

```python
#joining the tables and getting the final output
# SQL query to join crime data with population data
# SQL query to aggregate crimes by Area_Name (City), Group_Name (Crime Type), and Year
query_aggregate = """
    SELECT Area_Name AS City, Group_Name AS Crime_Type, Year, 
           SUM(Cases_Property_Stolen) AS Total_Crimes_Stolen,
           SUM(Cases_Property_Recovered) AS Total_Crimes_Recovered
    FROM crime_records
    GROUP BY Area_Name, Group_Name, Year
    ORDER BY Total_Crimes_Stolen DESC;
"""

# Execute query and store the result in a DataFrame
df_summary = pd.read_sql(query_aggregate, con=engine)

# Display the result
print(df_summary.head())
```

```python
#Assign Severity Scores to Different Crime Types
# SQL query to assign severity scores based on Group_Name (crime type)
query_severity_index = """
    SELECT Area_Name AS City, Year, Group_Name AS Crime_Type,
           CASE 
               WHEN Group_Name = 'Murder' THEN 10
               WHEN Group_Name = 'Rape' THEN 8
               WHEN Group_Name = 'Robbery' THEN 7
               WHEN Group_Name = 'Theft' THEN 3
               WHEN Group_Name = 'Assault' THEN 6
               ELSE 1
           END AS Severity_Score,
           SUM(Cases_Property_Stolen) AS Total_Crimes_Stolen,
           SUM(Cases_Property_Recovered) AS Total_Crimes_Recovered
    FROM crime_records
    GROUP BY Area_Name, Year, Group_Name
    ORDER BY Severity_Score DESC;
"""

# Execute query and store the result in a DataFrame
df_severity_index = pd.read_sql(query_severity_index, con=engine)

# Display the result
print(df_severity_index.head())
```

```python
# Calculate an Overall Severity Score for Each City
# SQL query to calculate an overall severity score for each city
query_overall_severity = """
    SELECT Area_Name AS City, Year, 
           SUM(CASE 
                   WHEN Group_Name = 'Murder' THEN 10
                   WHEN Group_Name = 'Rape' THEN 8
                   WHEN Group_Name = 'Robbery' THEN 7
                   WHEN Group_Name = 'Theft' THEN 3
                   WHEN Group_Name = 'Assault' THEN 6
                   ELSE 1
               END * Cases_Property_Stolen) AS Overall_Severity_Score
    FROM crime_records
    GROUP BY Area_Name, Year
    ORDER BY Overall_Severity_Score DESC;
"""

# Execute query and store the result in a DataFrame
df_overall_severity = pd.read_sql(query_overall_severity, con=engine)

# Display the result
print(df_overall_severity.head())
```

```python
# Rank Cities Based on Their Severity Scores
# SQL query to rank cities based on overall severity score
query_rank_cities = """
    SELECT Area_Name AS City, Year, 
           SUM(CASE 
                   WHEN Group_Name = 'Murder' THEN 10
                   WHEN Group_Name = 'Rape' THEN 8
                   WHEN Group_Name = 'Robbery' THEN 7
                   WHEN Group_Name = 'Theft' THEN 3
                   WHEN Group_Name = 'Assault' THEN 6
                   ELSE 1
               END * Cases_Property_Stolen) AS Overall_Severity_Score,
           RANK() OVER (ORDER BY SUM(CASE 
                                          WHEN Group_Name = 'Murder' THEN 10
                                          WHEN Group_Name = 'Rape' THEN 8
                                          WHEN Group_Name = 'Robbery' THEN 7
                                          WHEN Group_Name = 'Theft' THEN 3
                                          WHEN Group_Name = 'Assault' THEN 6
                                          ELSE 1
                                      END * Cases_Property_Stolen) DESC) AS Rank
    FROM crime_records
    GROUP BY Area_Name, Year
    ORDER BY Rank;
"""

# Execute query and store the result in a DataFrame
df_ranked_cities = pd.read_sql(query_rank_cities, con=engine)

# Display the result
print(df_ranked_cities.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Print the column names to verify
print(df.columns)

# Assuming the column is named 'Group_Name' (replace with the correct name if needed)
crime_stolen_by_group = df.groupby('group_name')['cases_property_stolen'].sum().reset_index()

# Plotting the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=crime_stolen_by_group, x='group_name', y='cases_property_stolen', palette='viridis')
plt.title('Total Crimes Stolen by Crime Type', fontsize=16)
plt.xlabel('Crime Type', fontsize=12)
plt.ylabel('Total Crimes Stolen', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'Year' and sum the 'Cases_Property_Stolen'
import matplotlib.pyplot as plt
import seaborn as sns
crime_stolen_by_year = df.groupby('year')['cases_property_stolen'].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=crime_stolen_by_year, x='year', y='cases_property_stolen', marker='o', color='blue')
plt.title('Trend of Property Crimes Stolen Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Property Crimes Stolen', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Correlation matrix
correlation_data = df[['cases_property_recovered', 'cases_property_stolen']].corr()

# Plotting the heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Between Property Recovered and Stolen', fontsize=16)
plt.show()
```

```python
# Count the occurrences of each 'Group_Name'
crime_type_counts = df['group_name'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(crime_type_counts, labels=crime_type_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3', len(crime_type_counts)))
plt.title('Proportion of Crime Types', fontsize=16)
plt.show()
```
